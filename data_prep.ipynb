```python
import pandas as pd
```

```python
pl23 = pd.read_csv('pl23-24.csv')
dfpl23 = pl23.iloc[:, 0:-82]
```

```python
#dfpl23 = dfpl23.drop('Div', axis=1)
dfpl23
```

```python
### Team-centric dataframe, extra column for points ###
dfpl23['Date'] = pd.to_datetime(dfpl23['Date'], dayfirst=True)
dfpl23 = dfpl23.sort_values(['Date', 'Time']).reset_index(drop=True)

# -------------------------
# HOME TEAM PERSPECTIVE
# -------------------------

home_df = pd.DataFrame({
    'Date': dfpl23['Date'],
    'Time': dfpl23['Time'],
    'Team': dfpl23['HomeTeam'],
    'Opponent': dfpl23['AwayTeam'],
    'Venue': 'H',

    'GoalsFor': dfpl23['FTHG'],
    'GoalsAgainst': dfpl23['FTAG'],

    'HTGoalsFor': dfpl23['HTHG'],
    'HTGoalsAgainst': dfpl23['HTAG'],

    'Shots': dfpl23['HS'],
    'ShotsAgainst': dfpl23['AS'],

    'ShotsOnTarget': dfpl23['HST'],
    'ShotsOnTargetAgainst': dfpl23['AST'],

    'Corners': dfpl23['HC'],
    'CornersAgainst': dfpl23['AC'],

    'Fouls': dfpl23['HF'],
    'FoulsAgainst': dfpl23['AF'],

    'YellowCards': dfpl23['HY'],
    'YellowCardsAgainst': dfpl23['AY'],

    'RedCards': dfpl23['HR'],
    'RedCardsAgainst': dfpl23['AR']
})

# Result from home team's perspective
home_df['Result'] = dfpl23['FTR'].map({
    'H': 'W',
    'D': 'D',
    'A': 'L'
})

# -------------------------
# AWAY TEAM PERSPECTIVE
# -------------------------

away_df = pd.DataFrame({
    'Date': dfpl23['Date'],
    'Time': dfpl23['Time'],
    'Team': dfpl23['AwayTeam'],
    'Opponent': dfpl23['HomeTeam'],
    'Venue': 'A',

    'GoalsFor': dfpl23['FTAG'],
    'GoalsAgainst': dfpl23['FTHG'],

    'HTGoalsFor': dfpl23['HTAG'],
    'HTGoalsAgainst': dfpl23['HTHG'],

    'Shots': dfpl23['AS'],
    'ShotsAgainst': dfpl23['HS'],

    'ShotsOnTarget': dfpl23['AST'],
    'ShotsOnTargetAgainst': dfpl23['HST'],
    

    'Corners': dfpl23['AC'],
    'CornersAgainst': dfpl23['HC'],

    'Fouls': dfpl23['AF'],
    'FoulsAgainst': dfpl23['HF'],

    'YellowCards': dfpl23['AY'],
    'YellowCardsAgainst': dfpl23['HY'],

    'RedCards': dfpl23['AR'],
    'RedCardsAgainst': dfpl23['HR']
})

# Result from away team's perspective
away_df['Result'] = dfpl23['FTR'].map({
    'A': 'W',
    'D': 'D',
    'H': 'L'
})

# -------------------------
# COMBINE
# -------------------------

team_df = pd.concat([home_df, away_df], ignore_index=True)

# Points column
team_df['Points'] = team_df['Result'].map({
    'W': 3,
    'D': 1,
    'L': 0
})

# Sort chronologically
team_df = team_df.sort_values(
    ['Team', 'Date', 'Time']
).reset_index(drop=True)

```

```python
### introduce positions and position difference features ###
dfpl23 = dfpl23.sort_values(['Date', 'Time']).reset_index(drop=True)

teams = sorted(
    set(dfpl23['HomeTeam']).union(set(dfpl23['AwayTeam']))
)

# League table
table = pd.DataFrame(index=teams)
table['Pts'] = 0
table['GD'] = 0
table['GF'] = 0

played = {team: 0 for team in teams}

home_pos = []
away_pos = []

for _, row in dfpl23.iterrows():

    home = row['HomeTeam']
    away = row['AwayTeam']

    # Current standings BEFORE this match
    standings = (
        table
        .sort_values(
            ['Pts', 'GD', 'GF'],
            ascending=False
        )
        .reset_index()
    )

    standings['Position'] = standings.index + 1

    pos_lookup = dict(
        zip(standings['index'], standings['Position'])
    )

    # First 3 matches => average position
    home_pos.append(
        10 if played[home] < 3 else pos_lookup[home]
    )

    away_pos.append(
        10 if played[away] < 3 else pos_lookup[away]
    )

    # Update table AFTER storing positions
    hg = row['FTHG']
    ag = row['FTAG']

    table.loc[home, 'GF'] += hg
    table.loc[home, 'GD'] += hg - ag

    table.loc[away, 'GF'] += ag
    table.loc[away, 'GD'] += ag - hg

    if hg > ag:
        table.loc[home, 'Pts'] += 3
    elif hg < ag:
        table.loc[away, 'Pts'] += 3
    else:
        table.loc[home, 'Pts'] += 1
        table.loc[away, 'Pts'] += 1

    played[home] += 1
    played[away] += 1

# Add to original match dataframe
dfpl23['HomeTablePos'] = home_pos
dfpl23['AwayTablePos'] = away_pos
dfpl23['TablePosDiff'] = (
    dfpl23['AwayTablePos']
    - dfpl23['HomeTablePos']
)
```

```python
# Home team rows
home_df['TeamTablePos'] = dfpl23['HomeTablePos']
home_df['OpponentTablePos'] = dfpl23['AwayTablePos']

# Away team rows
away_df['TeamTablePos'] = dfpl23['AwayTablePos']
away_df['OpponentTablePos'] = dfpl23['HomeTablePos']

# Difference from team's perspective
home_df['TablePosDiff'] = (
    home_df['OpponentTablePos']
    - home_df['TeamTablePos']
)

away_df['TablePosDiff'] = (
    away_df['OpponentTablePos']
    - away_df['TeamTablePos']
)
```

```python
### add the new features to the combined team dataframe ###
team_df = pd.concat([home_df, away_df], ignore_index=True)

team_df['Points'] = team_df['Result'].map({
    'W': 3,
    'D': 1,
    'L': 0
})

team_df = team_df.sort_values(
    ['Team', 'Date', 'Time']
).reset_index(drop=True)
```

```python
team_df
```

```python
### Compute rolling averages ###
rolling_cols = [
    'TablePosDiff',
    'GoalsFor',
    'Shots',
    'ShotsOnTarget',
    'Corners',
    'Fouls'
]

for col in rolling_cols:

    team_df[f'{col}_Rolling5'] = (
        team_df
        .groupby(['Team', 'Venue'])[col]
        .transform(
            lambda x: x.shift(1)
                      .rolling(window=5, min_periods=1)
                      .mean()
        )
    )

for col in rolling_cols:

    team_df[f'{col}_Rolling10'] = (
        team_df
        .groupby(['Team', 'Venue'])[col]
        .transform(
            lambda x: x.shift(1)
                      .rolling(window=10, min_periods=1)
                      .mean()
        )
    )

for col in rolling_cols:

    team_df[f'{col}_Rolling5_all'] = (
        team_df
        .groupby(['Team'])[col]
        .transform(
            lambda x: x.shift(1)
                      .rolling(window=5, min_periods=1)
                      .mean()
        )
    )

for col in rolling_cols:

    team_df[f'{col}_Rolling10_all'] = (
        team_df
        .groupby(['Team'])[col]
        .transform(
            lambda x: x.shift(1)
                      .rolling(window=10, min_periods=1)
                      .mean()
        )
    )
```

```python
home_features = team_df[team_df['Venue'] == 'H'].copy()
away_features = team_df[team_df['Venue'] == 'A'].copy()

home_features = home_features.add_prefix('Home_')
away_features = away_features.add_prefix('Away_')

home_features = home_features.rename(columns={
    'Home_Date': 'Date',
    'Home_Team': 'HomeTeam'
})

away_features = away_features.rename(columns={
    'Away_Date': 'Date',
    'Away_Team': 'AwayTeam'
})

matches = dfpl23.copy()

matches = matches.merge(
    home_features,
    on=['Date', 'HomeTeam'],
    how='left'
)

matches = matches.merge(
    away_features,
    on=['Date', 'AwayTeam'],
    how='left'
)

matches = matches.dropna()
```
